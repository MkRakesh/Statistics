# RandomForest_companydata

```python
Assignment
About the data: 
Let’s consider a Company dataset with around 10 variables and 400 records. 
The attributes are as follows: 
 Sales -- Unit sales (in thousands) at each location
 Competitor Price -- Price charged by competitor at each location
 Income -- Community income level (in thousands of dollars)
 Advertising -- Local advertising budget for company at each location (in thousands of dollars)
 Population -- Population size in region (in thousands)
 Price -- Price company charges for car seats at each site
 Shelf Location at stores -- A factor with levels Bad, Good and Medium indicating the quality of the shelving location for the car seats at each site
 Age -- Average age of the local population
 Education -- Education level at each location
 Urban -- A factor with levels No and Yes to indicate whether the store is in an urban or rural location
 US -- A factor with levels No and Yes to indicate whether the store is in the US or not
The company dataset looks like this: 
 
Problem Statement:
A cloth manufacturing company is interested to know about the segment or attributes causes high sale. 
Approach - A Random Forest can be built with target variable Sales (we will first convert it in categorical variable) & all other variable will be independent in the analysis.  
```

### Data Preprocessing

#### Scaling of continuous variable

```python
#MinMax Scaling
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
df.columns
```

```python
df.CompPrice=scaler.fit_transform(df[['CompPrice']])
df.Income=scaler.fit_transform(df[['Income']])
df.Advertising=scaler.fit_transform(df[['Advertising']])
df.CompPrice=scaler.fit_transform(df[['CompPrice']])
df.Price=scaler.fit_transform(df[['Price']])
df.Population=scaler.fit_transform(df[['Population']])
df.Age=scaler.fit_transform(df[['Age']])
df.Education=scaler.fit_transform(df[['Education']])
```

```python
df.tail()
```

#### OneHotEncoding

```python
# Let's encode categorical variables
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder(sparse_output=False, drop='first')  # drop='first' to avoid multicollinearity
df['ShelveLoc']= encoder.fit_transform(df[['ShelveLoc']])
df['Urban']= encoder.fit_transform(df[['Urban']])
df['US']= encoder.fit_transform(df[['US']])
df.tail()
```

#### Split Data into target and features

```python
#target
y=df.Sales
y.tail()
```

```python
#features
X=df.iloc[:,1:]
X.tail()
```

### RFC Modelling

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=10)
```

```python
#import RFC model
from sklearn.ensemble import RandomForestClassifier as RFC
```

#### 1.criterion='entropy'

```python
rfc_entropy=RFC(n_jobs=-1,n_estimators=10,criterion='entropy')
```

```python
rfc.fit(X_train,y_train)
```

#### Prediction and Evaluation

```python
y_pred1=rfc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

#Confusion Matrix
confusion_matrix(y_test,y_pred1)
```

```python
pd.crosstab(y_test,y_pred1)
```

```python
#Classification report
print(classification_report(y_test,y_pred1))
```

#### 2.criterion='gini'

```python
rfc_gini=RFC(n_jobs=-1,n_estimators=10,criterion='gini')
```

```python
rfc_gini.fit(X_train,y_train)
```

#### Prediction and Evaluation

```python
y_pred2=rfc_gini.predict(X_test)
```

```python
#Confusion Matrix
confusion_matrix(y_test,y_pred2)
```

```python
pd.crosstab(y_test,y_pred2)
```

```python
#Classification report
print(classification_report(y_test,y_pred2))
```

#### The model with entropy as criterion 1.gives slightly better prediction performance. 2. Better f1-score
           
#### Lets do some hypertuning 

### HyperTuning - GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
model=RFC(n_jobs=-1)
param_grid={'criterion':['entropy','gini'],'n_estimators':np.arange(1,100)}
grid_search=GridSearchCV(model,param_grid,cv=5)
```

```python
grid_search.fit(X_train,y_train)
```

```python
#best model
grid_search.best_estimator_
```

#### New model

```python
rfc_new=RFC(criterion='entropy', n_estimators=16, n_jobs=-1)
rfc_new.fit(X_train,y_train)
#prediction
y_pred3=rfc_new.predict(X_test)
#confusion matrix
confusion_matrix(y_test,y_pred3)
```

```python
pd.crosstab(y_test,y_pred3)
```

```python
#Confusion Matrix
print(classification_report(y_test,y_pred3))
```

#### ***precision ,recall and f1-score for both the classes have increased. Also accuracy increased by 2%

### Boosting -AdaBooster

```python
# AdaBoostClassifier:
from sklearn.ensemble import AdaBoostClassifier
# lets use our best model rfc_new as base model
rfc_boost = AdaBoostClassifier(estimator=rfc_new, n_estimators=2, random_state=22)
rfc_boost.fit(X_train, y_train)

# Prediction
y_pred4 = rfc_boost.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred4)
```

```python
pd.crosstab(y_test,y_pred4)
```

```python
#Confusion Matrix
print(classification_report(y_test,y_pred4))
```

#### 1.Significant improvement in f1-score for both class
#### 2.Improved precision ,recall and accuracy

### Error rate & Optimum value of k (n_estimators)

```python
error=[]
for k in range(1,100):
    rfc_= AdaBoostClassifier(estimator=rfc_new, n_estimators=k, random_state=22)
    rfc_.fit(X_train,y_train)
    y_pred5=rfc_.predict(X_test)
    error.append(np.mean(y_pred5!=y_test))
```

```python
#plot error vs different value of k
plt.figure(figsize=(12,4))
plt.grid()
plt.plot(range(1,100),error,marker='o',linestyle='dashed',markersize=5,color='b',markerfacecolor='r')
plt.xticks(np.linspace(0,100,26))
plt.xlabel('k_value')
plt.ylabel('Error')
plt.show()
```

#### *** Error rate is zero for k>1

### ROC and AUC

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
yt1=label_encoder.fit_transform(y_test)
yp1=label_encoder.fit_transform(y_pred4)

fpr,tpr,threshold =roc_curve(yt1,rfc_new.predict_proba(X_test)[:,1])
plt.subplots(figsize=(4,4))
plt.plot(fpr, tpr, color='blue',lw=1)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
```

```python
# Area under the curve
roc_auc_score(yt1, yp1)
```
