### Data Preprocessing

#### Categorize the target variable


```python
data['Taxable.Income']=data['Taxable.Income'].map(lambda x:'Risky' if x <= 30000 else 'Good')
data.head()
```

#### Min-Max Scaling

```python
from sklearn import preprocessing
# Use min-max scaling
scaler = preprocessing.MinMaxScaler()
data['City.Population']=scaler.fit_transform(data[['City.Population']])
data['Work.Experience']=scaler.fit_transform(data[['Work.Experience']])
data.head()
```

#### One-Hot-Encoding of Category variable

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False, drop='first')  # drop='first' to avoid multicollinearity
data['Undergrad']= ohe.fit_transform(data[['Undergrad']])
data['Marital.Status']= ohe.fit_transform(data[['Marital.Status']])
data['Urban']= ohe.fit_transform(data[['Urban']])
data.head() 
```

#### Split the data as target and feature variable

```python
y=data['Taxable.Income']
y.head()
```

```python
X=data.drop(['Taxable.Income'],axis=1)
X.head()
```

### RFC Modelling

```python
from sklearn.ensemble import RandomForestClassifier as RF
```

```python
# train test and split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=.25,random_state=11)
```

```python
#RFC Model 
rmodel=RF(n_jobs=-1,n_estimators=10,oob_score=True,criterion='entropy')
```

```python
rmodel.fit(X_train,y_train)
```

```python
#Accuracy
rmodel.score(X_test,y_test)
```

```python
#prediction
y_pred=rmodel.predict(X_test)
```

```python
#OuT of bag score (oob_score)
rmodel.oob_score_
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

#### Minority class prediction ability is low in our model. Lets model for a range of k values and remodel

#### Optimum value of k (n_estimators)

```python
error=[]
for k in range(5,100):
    rmodel1=RF(n_jobs=-1,n_estimators=k,oob_score=False,criterion='entropy')    
    rmodel1.fit(X_train,y_train)
    y_pred1=rmodel1.predict(X_test)
    error.append(np.mean(y_pred1!=y_test))
```

```python
#plot error vs different value of k
plt.figure(figsize=(12,4))
plt.grid()
plt.plot(range(5,100),error,marker='o',linestyle='dashed',markersize=5,color='b',markerfacecolor='r')
plt.xticks(np.linspace(0,100,26))
plt.xlabel('k_value')
plt.ylabel('Error')
plt.show()
```

#### Variation of error for 32>k<100 is minimal. Now lets do grid search too

### Grid-Search CV

```python
from sklearn.model_selection import GridSearchCV
grid={'n_estimators':np.arange(1,100)} # Parameter grid
rmodel_cv=RF(random_state=42)
grid_search = GridSearchCV(rmodel_cv,grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
```

```python
best_n_estimators = grid_search.best_params_['n_estimators']
print("Best n_estimators:", best_n_estimators)
```

#### New model with k=34

```python
rmodel_cv=RF(n_jobs=-1,n_estimators=34,oob_score=False,criterion='entropy')    
rmodel_cv.fit(X_train,y_train)
#prediction
y_pred_cv=rmodel_cv.predict(X_test)
```

```python
#Confusion matrix
confusion_matrix(y_test,y_pred_cv)
```

```python
#Classification Report
print(classification_report(y_test,y_pred_cv))
```
