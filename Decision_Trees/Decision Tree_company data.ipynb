#### Since our target variable is 'Sales', Let's covert it into classes as High & Low. Setting median of sales as threshold of low and high

```python
cdata1=cdata.copy()
cdata1['Sales']=cdata1['Sales'].map(lambda x:'low' if x < cdata1.Sales.median() else 'high')
```

### Data Processing

#### Let's do One-hot encoding for variables 'US','Urban' and 'ShelveLoc' also label encoding for 'Sales'

```python
# label_encoder = preprocessing.LabelEncoder()
# cdata1.Sales=label_encoder.fit_transform(cdata1['Sales'])
```

```python
from sklearn.preprocessing import OneHotEncoder

one_hot= OneHotEncoder(handle_unknown='ignore')
cdata2 = pd.DataFrame(one_hot.fit_transform(cdata1[['ShelveLoc','Urban','US']]).toarray())
```

```python
cdata2.head()
```

```python
cdata3=cdata1.drop(['ShelveLoc','US','Urban'],axis=1)
cdata3=pd.concat([cdata3,cdata2],axis=1)
cdata3.columns = cdata3.columns.astype(str)
cdata3.head()
```

##  1. Decision Tree Modelling (criterion='entropy')

```python
X=cdata3.iloc[:,1:]
y=cdata3.iloc[:,:1]
```

### train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2,random_state=44)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tmodel=DecisionTreeClassifier(criterion='entropy',max_depth=5)
tmodel.fit(X_train,y_train)
```

```python
print(cdata3.columns)
```

```python
 tree.plot_tree(tmodel);
```

```python
fn= ['CompPrice','Income', 'Advertising', 'Population', 'Price', 'Age', 'Education', '0', '1', '2', '3', '4', '5', '6']
cn=['low','high']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(tmodel,feature_names = fn, class_names=cn,filled=True);
```

### Prediction an Model evaluation

```python
y_pred=tmodel.predict(X_test)
```

```python
y_test1=np.array(y_test)
y_pred1=np.array(y_pred)
y_test1=y_test1.ravel()
```

```python
pd.crosstab(y_test1,y_pred1) # getting the 2 way table to understand the correct and wrong predictions
```

```python
tmodel.score(X_test,y_test)
```

#### Classificaton Report

```python
print(classification_report(y_test1,y_pred1))
```

#### ROC & AUC

```python
label_encoder = preprocessing.LabelEncoder()
yt=label_encoder.fit_transform(y_test1)
yp=label_encoder.fit_transform(y_pred1)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fpr,tpr,threshold = roc_curve(yt,tmodel.predict_proba(X_test)[:,1])
auc = roc_auc_score(yp, yt)
plt.subplots(figsize=(5,5))
plt.plot(fpr, tpr, color='b')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
```

## Decision Tree Modelling Gini impurity (criterion='gini')

```python
tmodel1=DecisionTreeClassifier(criterion='gini',max_depth=5)
```

```python
tmodel1.fit(X_train,y_train)
```

```python
fn= ['CompPrice','Income', 'Advertising', 'Population', 'Price', 'Age', 'Education', '0', '1', '2', '3', '4', '5', '6']
cn=['low','high']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(tmodel1,feature_names = fn, class_names=cn,filled=True);
```

### Prediction and Model evaluation

```python
y_pred2=tmodel1.predict(X_test)
```

```python
y_test2=np.array(y_test)
y_pred2=np.array(y_pred2)
y_test2=y_test2.ravel()
```

```python
pd.crosstab(y_test2,y_pred2) # getting the 2 way table to understand the correct and wrong predictions
```

```python
tmodel1.score(X_test,y_test)
```

#### Classificaton Report

```python
print(classification_report(y_test2,y_pred2))
```

```python
label_encoder = preprocessing.LabelEncoder()
yt1=label_encoder.fit_transform(y_test2)
yp1=label_encoder.fit_transform(y_pred2)
```

#### ROC & AUC

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fpr,tpr,threshold = roc_curve(yt,tmodel1.predict_proba(X_test)[:,1])
auc1 = roc_auc_score(yt1, yp1)
plt.subplots(figsize=(5,5))
plt.plot(fpr, tpr, color='b')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
```

```python
auc = roc_auc_score(yt1, yp1)
auc
```

### Conclusion

#### With a maximum depth of 5 the model based on 'entropy' gives better parameter compromisation than the one with 'gini impurity'.So tmodel1 can be considered as the better model.
