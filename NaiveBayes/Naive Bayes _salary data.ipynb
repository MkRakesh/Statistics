## Naive Bayes Classification

Since we have discretized the continous variables we can proceed to model Naive Bayes Classifier

```python
from sklearn.naive_bayes import CategoricalNB as CB
from sklearn.naive_bayes import MultinomialNB as MB
```

<b>Categorical NB

```python
classifier_cb = CB(alpha=True)
classifier_cb.fit(X_train,y_train)
pred_cb = classifier_cb.predict(X_test)
accuracy_cb= np.mean(pred_cb==y_test)
accuracy_cb
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, pred_cb)
```

```python
pd.crosstab(pred_cb,y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred_cb))
```

<b>Multinominal NB

```python
classifier_mb = MB()
classifier_mb.fit(X_train,y_train)
pred_mb = classifier_mb.predict(X_test)
accuracy_mb= np.mean(pred_mb==y_test)
accuracy_mb
```
