### MAPE metric funcion

```python
def MAPE(pred,original):
    mape=np.abs((pred-original)/original)*100
    return np.mean(mape)
```

```python
def MAPE(pred,org):
    temp = np.abs((pred-org)/org)*100
    return np.mean(temp)
```

# Forecasting Models

### 1. Simple Exponential

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
from sklearn.metrics import mean_squared_error
model_SE=SimpleExpSmoothing(df_train).fit(optimized=True)
pred_SE=model_SE.forecast(12)
RMSE_SE=np.sqrt(mean_squared_error(pred_SE,df_test.Passengers))
MAPE_SE=MAPE(pred_SE,df_test.Passengers)
```

```python
print('RMSE = ',RMSE_SE)
print('MAPE = ',MAPE_SE)
```

```python
plt.figure(figsize=(12,3.5))
plt.grid()
plt.plot(df_train,label='train data')
plt.plot(df_test,label='test data')
plt.plot(pred_SE,label='Simple Exponential Smoothing')
```

```python
df_train.isnull().sum()
```

### 2. Holt Winters double Exponential Smoothing

```python
from statsmodels.tsa.holtwinters import Holt
model_holts= Holt(df_train).fit(optimized=True) 
pred_holts=model_holts.forecast(12)
RMSE_holts=np.sqrt(mean_squared_error(pred_holts,df_test.Passengers))
MAPE_holts=MAPE(pred_holts,df_test.Passengers)
```

```python
print('RMSE = ',RMSE_holts)
print('MAPE = ',MAPE_holts)
```

## ARIMA

**Check for Stationarity**

```python
from statsmodels.tsa.stattools import adfuller
adf_test = adfuller(df_train)
print(f'p-value: {adf_test[1]}')
```

**p_value should be <.05 .lets do differencing**

```python
#Current value -previous value to find the new series
df_train_diff=df_train.diff().dropna()
plt.figure(figsize=(12,3))
plt.plot(df_train_diff.Passengers)
plt.title('1st order differenced data')
```

```python
from statsmodels.tsa.stattools import adfuller
adf_test1 = adfuller(df_train_diff)
print(f'p-value: {adf_test1[1]}')
```

**second order differencing**

```python
df_train_diff1=df_train_diff.diff().dropna()
plt.figure(figsize=(12,3))
plt.plot(df_train_diff1.Passengers)
plt.title('2st order differenced data')
```

```python
adf_test2 = adfuller(df_train_diff1)
print(f'p-value: {adf_test2[1]}')
```

**We made our series stattionary by second order differencing.  
Lets try some transformation**

**square root transformation**

```python
df1=np.sqrt(df_train.Passengers)
plt.figure(figsize=(12,3))
plt.plot(df1)
```

```python
adf_test3 = adfuller(df1)
print(f'p-value: {adf_test3[1]}')
```

**square root with differencing**

```python
df2=df1-df1.shift()
plt.figure(figsize=(12,3))
plt.plot(df2)
```

```python
adf_test4 = adfuller(df2[1:])
print(f'p-value: {adf_test4[1]}')
```

**log transformation**

```python
df3=np.log(df_train.Passengers)
plt.figure(figsize=(12,3))
plt.plot(df3)
```

```python
adf_test5 = adfuller(df3)
print(f'p-value: {adf_test5[1]}')
```

**log with differencing**

```python
df4=df3-df3.shift()
plt.figure(figsize=(12,3))
plt.plot(df4)
```

```python
adf_test6= adfuller(df2[1:])
print(f'p-value: {adf_test6[1]}')
```

**boxcox transformaton**

```python
from scipy.stats import boxcox
```

 ## Model based Methods

**Lets develope some ols forecasting models**

**Prepare data**

```python
data=air=pd.read_excel('Airlines+Data.xlsx')
```

```python
data.rename({'Month':'Date'},axis=1,inplace=True)
data.head()
```

```python
data['t'] = np.arange(1,len(data)+1)
data['t_squared'] = np.square(data.t)
```

```python
data["month"] = data.Date.dt.strftime("%b") # month extraction
data["year"] = data.Date.dt.strftime("%Y")  # year extraction
```

```python
months = pd.get_dummies(data['month'])
months = months[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']]
data = pd.concat([data,months],axis=1)
data['log_Passengers'] = np.log(data['Passengers'])
```

```python
data.head()
```

**Split the data**

```python
d_train=data[:84]
d_test=data[84:]
```

### OLS Models

```python
import statsmodels.formula.api as smf 
```

**1. Linear**

```python
model_linear = smf.ols('Passengers~t',data=d_train).fit()
pred_linear =  pd.Series(model_linear.predict(pd.DataFrame(d_test['t'])))
RMSE_linear = np.sqrt(np.mean((np.array(d_test['Passengers'])-np.array(pred_linear))**2))
RMSE_linear
```

**2. Exponential**

```python
model_exp= smf.ols('log_Passengers~t',data=d_train).fit()
pred_exp = pd.Series(model_exp.predict(pd.DataFrame(d_test['t'])))
RMSE_exp = np.sqrt(np.mean((np.array(d_test['Passengers'])-np.array(np.exp(pred_exp)))**2))
RMSE_exp
```

#### 3.Quadratic

```python
model_quad= smf.ols('Passengers~t+t_squared',data=d_train).fit()
pred_quad = pd.Series(model_quad.predict(pd.DataFrame(d_test[['t','t_squared']])))
RMSE_quad = np.sqrt(np.mean((np.array(d_test['Passengers'])-np.array(np.array(pred_quad)))**2))
RMSE_quad
```

**4. Additive**

```python
model_add= smf.ols('Passengers~Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov+Dec',data=d_train).fit()
pred_add = pd.Series(model_add.predict(d_test[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']]))
RMSE_add = np.sqrt(np.mean((np.array(d_test['Passengers'])-np.array(pred_add))**2))
RMSE_add
```

**5.Additive quadratic**

```python
model_add_quad= smf.ols('Passengers~t+t_squared+Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov+Dec',data=d_train).fit()
pred_add_quad = pd.Series(model_add_quad.predict(d_test[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','t','t_squared']]))
RMSE_add_quad= np.sqrt(np.mean((np.array(d_test['Passengers'])-np.array(pred_add_quad))**2))
RMSE_add_quad
```

**6. Multiplicative**

```python
model_mul= smf.ols('log_Passengers~Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov+Dec',data = d_train).fit()
pred_mul = pd.Series(model_mul.predict(d_test))
RMSE_mul = np.sqrt(np.mean((np.array(d_test['Passengers'])-np.array(np.exp(pred_mul)))**2))
RMSE_mul
```

#### 7. Multiplicative additive

```python
model_mul_add = smf.ols('log_Passengers~t+Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov+Dec',data = d_train).fit()
pred_mul_add = pd.Series(model_mul_add.predict(d_test))
rmse_mul_add = np.sqrt(np.mean((np.array(d_test['Passengers'])-np.array(np.exp(pred_mul_add)))**2))
rmse_mul_add
```

```python
MAPE_OLS=MAPE(np.exp(pred_mul_add),d_test.Passengers)
print('MAPE_OLS = ',MAPE(np.exp(pred_mul_add),d_test.Passengers))
```


**Among our model based methods, OLS Multiplicative additive model exhibitd least RMSE of 10.519. Lets visualize the prediction**

```python
plt.figure(figsize=(14,3))
plt.plot(d_train['Passengers'], label='Train')
plt.plot(d_test['Passengers'], label='Test')
plt.plot(np.exp(pred_mul_add), label='OLS Forecast')
plt.legend(loc='best')
plt.show()
```

**This model catches seasonality and trend very well**
