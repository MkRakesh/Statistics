### Data Preprocessing

```python
df['Quarter_Year'] = df['Quarter'].str.split('_').apply(lambda x: ' 19'.join(x[:]))
df.head()
```

```python
month=df['Quarter_Year'].str.split('_')
month

df['Month'] = df['Quarter_Year'].str.split(' ', expand=True)[0]
df['Year'] = df['Quarter_Year'].str.split(' ', expand=True)[1]

df['Month']=df['Month'].map({'Q1':'Jan','Q2':'Apr','Q3':'July','Q4':'Oct'})
df['Quarters']=df['Quarter_Year'].str.split(' ',expand=True)[0]
df.head()
```

```python
df['Date'] = (pd.to_datetime(df['Quarter_Year'].str.split(' ').apply(lambda x: ''.join(x[::-1])),dayfirst=True))
df.head()
```

## Forecasting Models

### 1. Simple Exponential Method

```python
# Function for MAPE
def MAPE(predicted,original):
    mape = np.abs((predicted-original)/original)*100
    return np.mean(mape)
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
```

```python
model_se= SimpleExpSmoothing(df_train.Sales).fit(optimized=True) 
pred_se=model_se.forecast(20)
pred_se.shape
```

```python
#metrics
from sklearn.metrics import mean_squared_error
rmse_se=np.sqrt(mean_squared_error(pred_se,df_test.Sales))
rmse_se
MAPE_se=MAPE(pred_se,df_test.Sales)
MAPE_se
print(' RMSE = ',rmse_se,'\n','MAPE = ',MAPE_se)
```

```python
plt.figure(figsize=(14,3.5))
plt.plot(df_train, label='Train')
plt.plot(df_test, label='Test')
plt.plot(pred_se, label='Simple exponential')
plt.legend(loc='best')
plt.title('Simple Exponential Forecasting')
plt.show()
```

### 2.Holts Method (Double exponential)

**Holts mothod captures both level and trend**

```python
from statsmodels.tsa.holtwinters import Holt
model_holts= Holt(df_train).fit(optimized=True) 
pred_holts=model_holts.forecast(20)
pred_holts.shape
```

```python
rmse_holts=np.sqrt(mean_squared_error(pred_holts,df_test.Sales))
MAPE_holts=MAPE(pred_holts,df_test.Sales)
print(' RMSE = ',rmse_holts,'\n','MAPE = ',MAPE_holts)
```

```python
plt.figure(figsize=(14,3.5))
plt.plot(df_train, label='Train')
plt.plot(df_test, label='Test')
plt.plot(pred_holts, label='Simple exponential')
plt.legend(loc='best')
plt.title('Simple Exponential Forecasting')
plt.show()
```

### Holt Winters Exponential Smoothing Method (Tripple Exponential)
#### Seasonality=Additive ,Trend=Additive



```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
model_holts1 =ExponentialSmoothing(df_train,seasonal='add',trend='add',seasonal_periods=12).fit(optimized=True)
pred_holts1 =model_holts1.forecast(20)
rmse_holts1=np.sqrt(mean_squared_error(pred_holts1,df_test.Sales))
MAPE_holts1=MAPE(pred_holts1,df_test.Sales)
print(' RMSE = ',rmse_holts1,'\n','MAPE = ',MAPE_holts1)
```

```python
plt.figure(figsize=(14,3.5))
plt.plot(df_train, label='Train')
plt.plot(df_test, label='Test')
plt.plot(pred_holts1, label='Simple exponential')
plt.legend(loc='best')
plt.title('Tripple Exponential Forecasting')
plt.show()
```

#### Seasonality=multiplicative ,Trend=multiplicative

```python
model_holts2 =ExponentialSmoothing(df_train,seasonal='mul',trend='mul',seasonal_periods=12).fit(optimized=True)
pred_holts2 =model_holts2.forecast(20)
rmse_holts2=np.sqrt(mean_squared_error(pred_holts2,df_test.Sales))
MAPE_holts2=MAPE(pred_holts2,df_test.Sales)
print(' RMSE = ',rmse_holts2,'\n','MAPE = ',MAPE_holts2)
```

```python
plt.figure(figsize=(14,3.5))
plt.plot(df_train, label='Train')
plt.plot(df_test, label='Test')
plt.plot(pred_holts2, label='Simple exponential')
plt.legend(loc='best')
plt.title('Tripple Exponential Forecasting')
plt.show()
```

#### Seasonality = additive ,Trend = multiplicative

```python
model_holts3 =ExponentialSmoothing(df_train,seasonal='add',trend='mul',seasonal_periods=12).fit(optimized=True)
pred_holts3 =model_holts3.forecast(20)
rmse_holts3=np.sqrt(mean_squared_error(pred_holts3,df_test.Sales))
MAPE_holts3=MAPE(pred_holts3,df_test.Sales)
print(' RMSE = ',rmse_holts3,'\n','MAPE = ',MAPE_holts3)
```

```python
plt.figure(figsize=(14,3.5))
plt.plot(df_train, label='Train')
plt.plot(df_test, label='Test')
plt.plot(pred_holts3, label='Simple exponential')
plt.legend(loc='best')
plt.title('Tripple Exponential Forecasting')
plt.show()
```

**Seasonality = multiplicative ,Trend = additive**

```python
model_holts4 =ExponentialSmoothing(df_train,seasonal='mul',trend='add',seasonal_periods=12).fit(optimized=True)
pred_holts4 =model_holts4.forecast(20)
rmse_holts4=np.sqrt(mean_squared_error(pred_holts4,df_test.Sales))
MAPE_holts4=MAPE(pred_holts4,df_test.Sales)
print(' RMSE = ',rmse_holts4,'\n','MAPE = ',MAPE_holts4)
```

## ARIMA Model

**Check for Stationarity**

#### Method 1. ACF & PACF Plots

```python
# with plt.rc_context():
plt.rc("figure", figsize=(14,3))
tsa_plots.plot_acf(df_train,lags=41)
plt.show()
plt.rc("figure", figsize=(14,3))
tsa_plots.plot_pacf(df_train,lags=41)
plt.show()
```

**Auto correlation is visible from the acf and pacf plots**

#### Method 2. AdFuller Test

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adf_test = adfuller(df_train)
print(f'p-value: {adf_test[1]}')
```

**p_value=0.99>0.05..Hence non-stationary**

### Transform to Stationary

#### Differencing

```python
#Current value -previous value to find the new series
df_train_diff=df_train.diff().dropna()
plt.plot(df_train_diff.Sales)
plt.title('1st order differenced data')
```

```python
adf_test1= adfuller(df_train_diff)
print(f'p-value: {adf_test1[1]}')
```

**The trend has vanished and lokks like mean is almost zero.Let's do AdFuller test again, But still it is non-satationary.Lets do differnce again**

```python
df_train_diff2=df_train_diff.diff().dropna()
plt.plot(df_train_diff2.Sales)
plt.title('2nd order differenced data')
```

```python
adf_test2= adfuller(df_train_diff2)
print(f'p-value: {adf_test2[1]}')
```

**p_value=0, showing strong indication of stationarity  
We can choose d=2 for our arima model**

### Selection of p & q values for ARIMA Model from ACF and PCF plots (Rule of Thumb)

```python
# with plt.rc_context():
plt.rc("figure", figsize=(14,3))
tsa_plots.plot_acf(df_train_diff2,lags=41)
plt.show()
plt.rc("figure", figsize=(14,3))
tsa_plots.plot_pacf(df_train_diff2,lags=41)
plt.show()
```

**In ACF plot peaks at 1,3,6,9,12 and almost decay from there .Lets choose an optimum value p=6  
In PACF plot it peaks at 3,6,9 and decays afterwards .Lets choose an optimum q of 6 **


```python
from statsmodels.tsa.arima.model import ARIMA
model_ARIMA=ARIMA((df_train.Sales),order=(6,2,9)).fit()
pred_ARIMA=model_ARIMA.forecast(20)
```

```python
MAPE_ARIMA = MAPE(pred_ARIMA,df_test.Sales)
rmse_ARIMA = np.sqrt(mean_squared_error(pred_ARIMA,df_test.Sales))
print('MAPE = ',MAPE(pred_ARIMA,df_test.Sales))
print('rmse = ',np.sqrt(mean_squared_error(pred_ARIMA,df_test.Sales)))
```

```python
plt.figure(figsize=(14,3.5))
plt.plot(df_train, label='Train')
plt.plot(df_test, label='Test')
plt.plot(pred_ARIMA, label='ARIMA')
plt.legend(loc='best')
plt.title('ARIMA Forecasting')
plt.show()
```

**The ARIMA model exhibits satisfactory MAPE and rmse  
Also it catches seasonality and trend**

## SARIMA

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
model_SARIMA = SARIMAX(df_train, order=(3,1,3), seasonal_order=(1,1,1,12)).fit()
pred_SARIMA= model_SARIMA.forecast(20)
```

```python
MAPE_SARIMA =MAPE(pred_SARIMA,df_test.Sales)
rmse_SARIMA = np.sqrt(mean_squared_error(pred_SARIMA,df_test.Sales))
print('MAPE = ',MAPE(pred_SARIMA,df_test.Sales))
print('rmse = ',np.sqrt(mean_squared_error(pred_SARIMA,df_test.Sales)))
```

```python
plt.figure(figsize=(14,3.5))
plt.plot(df_train, label='Train')
plt.plot(df_test, label='Test')
plt.plot(pred_SARIMA, label='SARIMA')
plt.legend(loc='best')
plt.title('SARIMA Forecasting')
plt.show()
```

**SARIMA Model gives appreciable MAPE and RMSE value  
Also catches seasonality and trend**
