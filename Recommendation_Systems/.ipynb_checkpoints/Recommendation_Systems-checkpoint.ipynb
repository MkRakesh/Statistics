**Lets create a pivot table from our data**

```python
df.drop_duplicates(subset=['User.ID', 'Book.Title'], keep='first', inplace=True)
```

```python
books_df = df.pivot(index='User.ID',
                                 columns='Book.Title',
                                 values='Book.Rating').reset_index(drop=True)
```

```python
books_df.head()
```

```python
books_df.index=df['User.ID'].unique()
```

```python
books_df.tail()
```

```python
books_df.fillna(0,inplace=True)
books_df.drop(books_df.columns[-1],inplace=True,axis=1)
```

```python
books_df.head()
```

**Calculate Cosine similarity**

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
#Calculating Cosine Similarity between Users
books_sim = 1 - pairwise_distances( books_df.values,metric='cosine')
```

```python
books_sim
```

```python
books_sim_df=pd.DataFrame(books_sim)
books_sim_df.head()
```

```python
#Set the index and column names to user ids 
books_sim_df.index = df['User.ID'].unique()
books_sim_df.columns = df['User.ID'].unique()
books_sim_df
```

```python
np.fill_diagonal(books_sim, 0)
books_sim_df.iloc[0:5, 0:5]
```

**Most Similar Users**

```python
books_sim_df.idxmax(axis=1)[:5]
```
