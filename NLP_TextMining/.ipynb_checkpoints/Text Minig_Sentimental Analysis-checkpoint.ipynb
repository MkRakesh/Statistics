Sentimental Analysis is composed of  extracting subjectivity and polarity of the individual text data (tweets in our case) and analyse the sentiments based on subjectivity and polarity score of each record. It can be done mainly in two ways<br>
<br>1.Using TextBlob<br>
<br>2.Using NLTK's SentimentIntensityAnalyzer (SIA)<br>  
<br> Let's do some data preprocessing before that


### Data preprocessing

```python
df_tweet=data.copy()
df_tweet.drop('Unnamed: 0',inplace=True,axis=1)
df_tweet.rename({'Text':'tweets'},axis=1,inplace=True)
df_tweet.head()
```

 <br>We can see lot of junks in the tweets such as urls, tags, RT and  Hashtags. let's remove them using re library

```python
import re 
```

```python
# function to clean junks

def clean_tweets(tweets): 
    tweets = re.sub(r'@\w+','',tweets)            # Removes @name
    tweets = re.sub('#','',tweets)                # Removes # (Hashtags)
    tweets = re.sub(r'^RT\s+@\w+:\s+','',tweets)  # Removes RT (Retweets)
    tweets = re.sub('https?:\/\/\S+', '', tweets) # Removes Links
    tweets = re.sub(r'<U\+[0-9A-F]+>', '', tweets)# Removes Unicode escape sequences
    
    return tweets
```

```python
# Apply the function to clean the texts
df_tweet['tweets']=df_tweet['tweets'].apply(clean_tweets)

# To see the full text
pd.set_option("display.max_colwidth", None)
df_tweet.head()
```

### 1. Sentimental analysis using TextBlob

```python
# !pip install textblob
```

```python
from textblob import TextBlob
```

TextBlob return subjectivity(0 to -1) and polarity(-1 to 1)

```python
#function to return subjecticity and polarity
def get_polarity(tweets):
    return TextBlob(tweets).sentiment.polarity
def get_subjectivity(tweets):
    return TextBlob(tweets).sentiment.subjectivity

df_tweet['subjectivity_score']=df_tweet.tweets.apply(get_subjectivity)
df_tweet['polarity_score']=df_tweet.tweets.apply(get_polarity)
df_tweet.head()
```

```python
df_tweet['subjectivity']=df_tweet['subjectivity_score'].map(lambda x: 'neg' if x<0 else 'pos' if x>0 else 'neutral')
df_tweet['polarity']=df_tweet['polarity_score'].map(lambda x: 'neg' if x<0 else 'pos' if x>0 else 'neutral')
df_tweet.tail()
```

```python
plt.figure(figsize=(4,3.5))
plt.grid()
plt.hist(df_tweet.iloc[:,-1],label='polarity')
plt.hist(df_tweet.iloc[:,-2],label='subjectivity')
plt.legend(loc='best')
plt.show()
```

<br>We can see thet majority of the tweets are recognized as neutral and positive .Lets do a scatter plot on it

```python
import seaborn as sns
sns.scatterplot(data=df_tweet,x='polarity_score',y='subjectivity_score',hue='polarity')
```

### 2.  Sentimental analysis using NLTK- SIA

```python
import nltk
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

Let's see how SIA works with a sentence

```python
sia= SentimentIntensityAnalyzer()
sia.polarity_scores('I love beaches')
```

It returns negative,positive,neutral and a compound score<br>
<br>Let's go ahead and apply to the tweets

```python
#function to return the polarity score
def sentiment_SIA(tweets):
    sia = SentimentIntensityAnalyzer()
    polarity_scores = sia.polarity_scores(tweets)
    return 'neg' if polarity_scores['compound'] <0  else 'neu' if polarity_scores['compound']==0 else 'pos'
```

```python
df_tweet['polarity_sia']=df_tweet.tweets.apply(lambda tweets: sentiment_SIA(tweets))
```

```python
df_tweet.tail()
```

Lets create a barplot and piechart on poliarity obtained using SIA

```python
plt.figure(figsize=(4,3.5))
plt.grid()
plt.hist(df_tweet.iloc[:,-1],label='polarity_sia')
plt.legend(loc='best')
plt.show()
```

Almost 90% tweets are treated as positive

```python
data1=df_tweet.polarity_sia.value_counts()
plt.pie(x=data1, labels = data1.index, autopct = '%.0f%%',colors=sns.color_palette('pastel'));
```

We get 44% of positive and neutral tweets and 12% negative with SIA
